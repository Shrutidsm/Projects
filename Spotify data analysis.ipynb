```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks=pd.read_csv("tracks.csv")
df_tracks.head()
```

```python
pd.isnull(df_tracks).sum()
```

```python
df_tracks.info()
```

```python
## show the most and least popular songs

sorted_df=df_tracks.sort_values('popularity').tail(10)
sorted_df
```

```python
df_tracks.describe().transpose()
#transpose function is used to write rows as columns and columns as rows
```

```python
## display the top 10 popular songs from the dataset
most_popular=df_tracks.query('popularity>90', inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
df_tracks
```

```python
## change my dataset index to release date and change the format of the release to datetime
df_tracks.set_index("release_date",inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index) ##
df_tracks
```

```python
df_tracks[['artists']].iloc[18]
```

```python
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x:round(x/1000))
df_tracks.drop('duration_ms',inplace=True,axis=1)
```

```python
df_tracks.duration.head()
```

```python
df_tracks.head()
```

```python
corr_df=df_tracks.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=" .2g",vmin=-1,vmax=1,center=0,cmap='ocean',linewidth=1,linecolor='Black')
heatmap.set_title("Correlation HeatMap Between variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(14,7))
sns.regplot(data=sample_df,y='loudness',x='energy',color='c').set(title='Loudness Vs Energy Correlation plot')
```

```python
plt.figure(figsize=(14,7))
sns.regplot(data=sample_df, y='popularity',x='acousticness',color='r').set(title="Popularity Vs Acousticness Correlation plot")
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title='Number of songs per year')
```

```python
total_dr=df_tracks.duration
fig_dims=(18,7)
fig,ax = plt.subplots(figsize=fig_dims) ##
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title='Years vs Duartion')
plt.xticks(rotation=90)
```

```python
total_dr=df_tracks.duration
sns.set_style(style='whitegrid')
fig_dims=(10,5)
fig,ax = plt.subplots(figsize=fig_dims) ##
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title='Years vs Duartion')
plt.xticks(rotation=90)
```

```python
df_genre=pd.read_csv('SpotifyFeatures.csv')
df_genre.head()
```

```python
plt.title('Duration of songs in different genres')
plt.figure(figsize=(20,10))
sns.color_palette('twilight',as_cmap=True)
sns.barplot(data=df_genre,x='duration_ms',y='genre')
plt.xlabel('Duration in milli seconds')
plt.ylabel('Genre')
```

```python
sns.set_style(style='darkgrid')
plt.figure(figsize=(20,8))
famous=df_genre.sort_values('popularity',ascending=False).head(10)
sns.barplot(data=famous, x='popularity',y='genre').set(title='Top 5 genres by Popularity')
```
